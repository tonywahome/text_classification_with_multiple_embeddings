# TF-IDF + Dense Neural Network
# Text Classification on Amazon Reviews

This notebook implements text classification using **TF-IDF** features with a **Dense Neural Network**.

**Approach**: TF-IDF (Term Frequency-Inverse Document Frequency)
- Statistical representation of text
- Captures term importance across documents
- No semantic understanding
- Fast and interpretable

**Model**: Deep Neural Network (not GRU, since TF-IDF is not sequential)
- Multiple dense layers
- Dropout for regularization
- Suitable for fixed-size feature vectors
